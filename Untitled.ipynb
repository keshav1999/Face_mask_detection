```python
from keras.models import load_model
import cv2
import numpy as np
import tkinter
from tkinter import messagebox
import smtplib

# Initialize Tkinter
root = tkinter.Tk()
root.withdraw()

#Load trained deep learning model
model = load_model('Final.h5')

#Classifier to detect face
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
face_det_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_det_classifier.load("'haarcascade_frontalface_default.xml'")
print(face_det_classifier)

# Capture Video
cap=cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if ret == False:
        continue
    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect multi faces in the image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for face in faces:
        x, y, w, h = face

        # Get the face ROI
        offset = 7
        face_section = frame[y-offset:y+h+offset, x-offset:x+w+offset]
        face_section = cv2.cvtColor(face_section, cv2.COLOR_BGR2GRAY)
        print(face_section.shape)
        face_section = cv2.resize(face_section, (100, 100))
        print(face_section.shape) 
        face_section = face_section.reshape((1,100,100,1))
        print(face_section.shape)
        out = model.predict(face_section)
        print(out)

        # Draw rectangle in the original image
        cv2.putText(frame, str(out),(x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 1,(255,0,0),2,cv2.LINE_AA)
        cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)

    cv2.imshow("Faces", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
